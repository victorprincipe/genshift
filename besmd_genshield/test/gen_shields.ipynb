```python
def bohr_2_ang(file):
    bohr = read(file, index=":")

    for frame in bohr:
        frame.cell *= ase.units.Bohr
        frame.positions *= ase.units.Bohr

    write(file[:-4]+'_ang.xyz', bohr)
    print("Conversion complete. File saved as: "+file[:-4]+'_ang.xyz' )
```

```python
bohr_2_ang('glycine_short.xyz')
```
